```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("ipdatarecord-txrx-nonzero-raemis.csv")
print(df)
```

```python
print(df.columns)
```

```python
print(df.head())
```

```python
df_october = df.loc[(df['start_time'] >= '2025-10-01')
                     & (df['start_time'] < '2025-11-1')]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_october['start_time'], df_october['tx_total_packets'] + df_october['rx_total_packets'], alpha=0.5)
plt.xlabel('Start Time')
plt.ylabel('Transmitted and Recieved Packets')
plt.title('Usage over Time')
plt.grid(False)
plt.show()
```

```python
df_october_nonzero = df_october.loc[(df_october['tx_total_packets'] != 0)
                     & (df_october['rx_total_packets'] != 0)]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_october_nonzero['start_time'], df_october_nonzero['tx_total_packets'] + df_october_nonzero['rx_total_packets'], alpha=0.5)
plt.xlabel('Start Time')
plt.ylabel('Transmitted and Recieved Packets')
plt.title('Usage over Time')
plt.grid(False)
plt.show()
```
