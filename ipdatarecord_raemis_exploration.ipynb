```python
df['start_time'] = pd.to_datetime(df['start_time'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['start_time'], df['tx_total_packets'] + df['rx_total_packets'], alpha=0.5, s=6)
plt.xlabel('Start Time')
plt.ylabel('Transmitted and Recieved Packets')
plt.title('Usage over Time')
plt.grid(False)
plt.show()
```

```python
df['tx_and_rx_mb'] = (df['tx_total_packets'] + df['rx_total_packets']) * 1500/1000000
```

```python
df_school_year = df.loc[((df['start_time'] >= '2024-9-03') & (df['start_time'] < '2025-6-18'))
                     | (df['start_time'] >= '2025-9-03')]
df_summer_break = df.loc[(df['start_time'] < '2024-9-03') |
                     ((df['start_time'] >= '2025-6-18')
                     & (df['start_time'] < '2025-9-03'))]
```

```python
print(len(df))
print(len(df_school_year))
print(len(df_summer_break))
plt.figure(figsize=(10, 6))
plt.scatter(df_summer_break['start_time'], df_summer_break['tx_and_rx_mb'], alpha=0.5, s=6)
plt.xlabel('Start Time')
plt.ylabel('Transmitted and Recieved Megabytes')
plt.title('Usage over Summer Break')
plt.grid(False)
plt.show()
```

```python
df_school_weekdays = df_school_year.loc[df_school_year['start_time'].dt.dayofweek < 5]
df_school_weekends = df_school_year.loc[df_school_year['start_time'].dt.dayofweek >= 5]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_school_weekends['start_time'], df_school_weekends['tx_and_rx_mb'], alpha=0.5, s=6)
plt.xlabel('Start Time')
plt.ylabel('Transmitted and Recieved Megabytes)
plt.title('Usage over Weekends in the School Year')
plt.grid(False)
plt.show()
```

```python
df_school_weekdays['hour'] = df_school_weekdays['start_time'].dt.hour
school_weekday_hourly_avg = df_school_weekdays.groupby('hour')['tx_and_rx_mb'].agg(['mean', 'sum', 'count', 'std'])
school_weekday_hourly_avg = school_weekday_hourly_avg.reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(school_weekday_hourly_avg['hour'], school_weekday_hourly_avg['mean'])
plt.xlabel('Start Time')
plt.ylabel('Transmitted and Recieved Megabytes')
plt.title('Average Usage by Hour of School Weekday')
plt.xticks(range(0, 24), [f'{h:02d}:00' for h in range(24)], rotation=45, ha='right')
plt.grid(False)
plt.show()
# average doesn't tell us much, peaks are sparse and not regular
```

```python
plt.figure(figsize=(10, 6))
plt.bar(school_weekday_hourly_avg['hour'], school_weekday_hourly_avg['sum'])
plt.xlabel('Start Time')
plt.ylabel('Transmitted and Recieved Megabytes')
plt.title('Sum of Usage by Hour of Weekday')
plt.xticks(range(0, 24), [f'{h:02d}:00' for h in range(24)], rotation=45, ha='right')
plt.grid(False)
plt.show()
# there are high outliers in early morning skewing the average
# should i remove these outliers to see a more accurate average day?
```

```python
df_summer_break['hour'] = df_summer_break['start_time'].dt.hour
summer_break_hourly_avg = df_summer_break.groupby('hour')['tx_and_rx_mb'].agg(['mean', 'sum', 'count', 'std'])
summer_break_hourly_avg = summer_break_hourly_avg.reset_index()
```

```python
school_weekday_hourly_users = df_school_weekdays.groupby(['start_time', 'hour'])['sub_ref_1'].nunique().reset_index()
school_weekday_hourly_users.columns = ['start_time', 'hour', 'unique_count']

# Now get statistics of those unique counts by hour
school_weekday_hourly_users = school_weekday_hourly_users.groupby('hour')['unique_count'].agg([
    'mean',      # Average number of unique values
    'median',    # Median
    'std',       # Standard deviation
    'min',       # Minimum
    'max',       # Maximum
    'count'      # Number of days
]).reset_index()
```
